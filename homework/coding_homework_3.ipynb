```python
# Importing important packages

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Make a Code Cell that imports your v_rest_sst.csv and v_rest_pvalb.csv variables as numpy arrays called v_rest_sst and v_rest_pvalb respectively.
sst_url = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/v_rest_sst.csv'
v_rest_sst = np.loadtxt(sst_url)

pvalb_url = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt(pvalb_url)

# verification of array type
type(v_rest_sst)
```

```python
# Finds means and performs t-test
# Depending on the t-test result print histograms

import statistics as st #using stats package to find mean
v_rest_sst_mean = st.mean(v_rest_sst)
v_rest_pvalb_mean = st.mean(v_rest_pvalb)
print(v_rest_sst_mean)
print(v_rest_pvalb_mean)

ttest_result = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
print(ttest_result)

if ttest_result.pvalue > 0.05:
  fig, (ax1, ax2) = plt.subplots(1, 2) #creates side by side subplots
  ax1.hist(v_rest_sst)
  ax1.set_title('sst')
  ax1.set_xlabel('resting membrane potential (mV)')
  ax1.set_ylabel('frequency')
  ax2.hist(v_rest_pvalb)
  ax2.set_title('pvalb')
  ax2.set_xlabel('resting membrane potential (mV)')
  ax2.set_ylabel('frequency')
```

```python
# creating function for above code

def vrestHist(data1, data2): #input two datasets so it can be used for different data
  v_rest_sst_mean = st.mean(data1)
  v_rest_pvalb_mean = st.mean(data2)
  print(v_rest_sst_mean)
  print(v_rest_pvalb_mean)

  ttest_result = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
  print(ttest_result)

  if ttest_result.pvalue > 0.05:
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.hist(v_rest_sst)
    ax1.set_title('sst')
    ax1.set_xlabel('resting membrane potential (mV)')
    ax1.set_ylabel('frequency')
    ax2.hist(v_rest_pvalb)
    ax2.set_title('pvalb')
    ax2.set_xlabel('resting membrane potential (mV)')
    ax2.set_ylabel('frequency')
```

```python
# calling function with numpy arrays as inputs

vrestHist(v_rest_sst, v_rest_pvalb)
```

```python
# color coded scatter plot for sst versus pvalb resting membrane potentials
for i in range(len(v_rest_sst)):
  if v_rest_sst[i]>-60 and v_rest_pvalb[i]>-60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'red')
  if v_rest_sst[i]>-60 and v_rest_pvalb[i]<-60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'green')
  if v_rest_sst[i]<-60 and v_rest_pvalb[i]<-60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'blue')
  if v_rest_sst[i]<-60 and v_rest_pvalb[i]>-60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'purple')
#labeling plot
plt.title('sst vs. pvalb')
plt.xlabel('sst')
plt.ylabel('pvalb')
```

```python
# create a function to generate scatter plot
def vrestScatter(data1, data2):
  # save datasets as new variables
  v_rest_sst = data1
  v_rest_pvalb = data2
  for i in range(len(v_rest_sst)):
    if v_rest_sst[i]>-60 and v_rest_pvalb[i]>-60:
      plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'red')
    if v_rest_sst[i]>-60 and v_rest_pvalb[i]<-60:
      plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'green')
    if v_rest_sst[i]<-60 and v_rest_pvalb[i]<-60:
      plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'blue')
    if v_rest_sst[i]<-60 and v_rest_pvalb[i]>-60:
      plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color = 'purple')
  plt.title('sst vs. pvalb')
  plt.xlabel('sst')
  plt.ylabel('pvalb')
```

```python
# verification of function with inputs
vrestScatter(v_rest_sst, v_rest_pvalb)
```
