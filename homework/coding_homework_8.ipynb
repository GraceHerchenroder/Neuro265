```python
# Import relevant details

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Upload relevant files

url_ext = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/extinction_predictors_hw8.csv'
url_fear = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/fear_conditioning_traces_hw8.csv'

extinction_predictors = np.loadtxt(url_ext, delimiter = ',')
fear_traces = np.loadtxt(url_fear, delimiter = ',')

shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int)
```

```python
# Separating fear_traces by conditioning vs extinction

conditioning_traces = fear_traces[:10458, :]
extinction_traces = fear_traces[10459:, :]
```

```python
# Printing size and shape of new arrays

print('Conditioning - ', 'Size:', conditioning_traces.size, 'Shape:', conditioning_traces.shape)
print('Extinction - ', 'Size:', extinction_traces.size, 'Shape:', extinction_traces.shape)
```

```python
# Creating time arrays for both conditioning and extinction

frame_rate = 15

dur_cond = (conditioning_traces.shape[0])/frame_rate
dur_ext = extinction_traces.shape[0]/frame_rate
print('seconds: ',dur_cond, ',',dur_ext)
print('minutes: ',dur_cond/60,',', dur_ext/60)

time_conditioning = np.linspace(0, dur_cond, conditioning_traces.shape[0])
time_extinction = np.linspace(0, dur_ext, extinction_traces.shape[0])
```

There are **63** ROIs (cells) that were imaged during the sessions.

The total length of time for each:
- **Conditioning**: 697.2 seconds or 11.62 minutes
- **Extinction**: 947.267 seconds or 15.79 minutes

```python
# Plot the first and second ROIs for each case in subplots
fig, axs = plt.subplots(2)
axs[0].plot(time_conditioning, conditioning_traces[:,0], 'b')
axs[0].plot(time_conditioning, conditioning_traces[:,1], 'r')
axs[0].vlines(shock_index, 0, 1, 'k', '--')
axs[0].vlines(tone_index, 0, 1, 'g', '--')
axs[0].set_xlabel('time [s]')
axs[0].set_ylabel('trace ($\Delta$F/F)')
axs[0].set_title('Conditioning')

axs[1].plot(time_extinction, extinction_traces[:,0], 'b')
axs[1].plot(time_extinction, extinction_traces[:,1], 'r')
axs[1].set_xlabel('time [s]')
axs[1].set_ylabel('trace ($\Delta$F/F)')
axs[1].set_title('Extinction')

fig.legend(['ROI 0', 'ROI 1'])
```

The average trace value for both conditioning and extinction is about the same before and after the shock/tone at about 0.4 F/F. However, the extinction has more noise in the resulting data.

For the first ROI, there does not appear to be any correlation between the conditioning and extinction. For the conditioning, the trace peaks at the shock each time. However, for extinction it appears as though the trace is higher overall. My guess would be that the mouse is more on edge waiting for the tone and shock combination.

For the second ROI, the overall noise of the data appears to be less. But, there is more of a correlation between the conditioning and extinction sets. The initial trace peak after the shock occurs in both the conditioning and the extinction ROIs. Due to this, I would say that there is some relationship between how a cell responds during conditioning and extinction as the trace peaks at the same location.

If not a relationship throughout the extinction process, the extinction is affected by the work done during connection which is evident from the increased trace values overall in ROI 0.

```python
# Create a predictors variable for modelling purposes

# Import relevant files for doing so
from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array

print(predictors_conditioning)
```

```python
# Correlating the predicted trace and the actual trace using r for correlation
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model_results = model.fit()
  predicted_trace = model_results.predict()
  r = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r.statistic

plt.plot(np.sort(r_model_conditioning), np.linspace(0, 1, len(r_model_conditioning), endpoint=False), label = "model")
plt.legend()
plt.xlabel('Correlation Between Model Prediction and Observed Trace')
plt.ylabel('Cumulative Proportion of ROIs')
```

```python
# Correlating the predicted trace and actual trace for extinction

tone_array = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  model_results = model.fit()
  predicted_trace = model_results.predict()
  r = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r.statistic

plt.plot(np.sort(r_model_extinction), np.linspace(0, 1, len(r_model_extinction), endpoint=False), label = "model")
plt.legend()
plt.xlabel('Correlation Between Model Prediction and Observed Trace')
plt.ylabel('Cumulative Proportion of ROIs')
```

```python
# Correlating r for conditioning and exctinction

pearsonr(r_model_conditioning, r_model_extinction)
```

```python
# Making a scatter plot with r conditioning and r extinction

plt.scatter(r_model_conditioning, r_model_extinction)
plt.xlabel('Conditioning')
plt.ylabel('Extinction')
```

According to the correlation coefficient, the conditioning and extinction traces are not correlated as the p-value is greater than 0.05. This makes intuitive sense, as over time for extinction the mouse would get used to the sound without the shock.

However, based off of what I understand from the scatter plot, as the conditioning r value gets higher so does this extinction r value. I assume that this indicates that with stronger response during the correlation the higher the likelihood that there will be an affect during the extinction. Though for the most part the scatter plot looks arbitrary with no clear trends.
