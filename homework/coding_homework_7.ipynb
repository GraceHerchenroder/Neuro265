```python
# Import relevant modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import data into numpy arrays using delimiter
url_right1 = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/spikes_right_trial1.csv'
url_right2 = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/spikes_right_trial2.csv'
url_left1 = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/spikes_left_trial1.csv'
url_left2 = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/spikes_left_trial2.csv'

spikes_left_trial1 = np.loadtxt(url_left1, delimiter = ',')
spikes_left_trial2 = np.loadtxt(url_left2, delimiter = ',')
spikes_right_trial1 = np.loadtxt(url_right1, delimiter = ',')
spikes_right_trial2 = np.loadtxt(url_right2, delimiter = ',')
```

```python
# Print length of each array
# Save in variables
len_left1 = len(spikes_left_trial1)
len_left2 = len(spikes_left_trial2)
len_right1 = len(spikes_right_trial1)
len_right2 = len(spikes_right_trial2)

print('Left Trial 1:',len_left1)
print('Left Trial 2:',len_left2)
print('Right Trial 1:',len_right1)
print('Right Trial 2:',len_right2)
```

```python
# Firing rate of the neuron during each trial
duration = 30 # 30-second waiting period

fr_left1 = len_left1/duration
fr_left2 = len_left2/duration
fr_right1 = len_right1/duration
fr_right2 = len_right2/duration

print('Left Trial 1:',fr_left1, 'Hz')
print('Left Trial 2:',fr_left2, 'Hz')
print('Right Trial 1:',fr_right1, 'Hz')
print('Right Trial 2:',fr_right2, 'Hz')
```

Considering the firing rates shown above, it does not appear as though the neuron fires more during either left or right turn trials. Overall, all trials for both left and right display firing rates less than 2.5 Hz. This seems to be low in comparison to firing rates found during in class examples (see 'Spikes_04_01_24'). This means that for both left and right trials there are betwwen 1.5 and 2.5 spikes per second over the given time period.

The frequencies are closer together/more consistent for the left neuron meaning that it may be more reliable to think the neuron would fire more during left turns as the firing rate is centered about 2 Hz while for right turns it goes both lower and higher. However, this is an assumption drawn on two data points for the left turns which is not enough to draw an actual conclusion on consistency.

```python
# Plotting spike train data for trial 1 and trial 2
f, a = plt.subplots(2, 1, figsize = (8,10))

# Trial 1
a[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
a[0].plot(spikes_right_trial1, 2*np.ones_like(spikes_right_trial1), '.')
a[0].set_ylim([0, 3])
a[0].set_yticks([])
a[0].set_xlabel('time (s)')
a[0].set_title('Trial One')
a[0].legend(['Left', 'Right'])

# Trial 2
a[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
a[1].plot(spikes_right_trial2, 2*np.ones_like(spikes_right_trial2), '.')
a[1].set_ylim([0, 3])
a[1].set_yticks([])
a[1].set_xlabel('time (s)')
a[1].set_title('Trial Two')
a[1].legend(['Left', 'Right'])
```

### Comparing Left and Right Turns

The spike train data indicated the times at which the neuron is firing. From this, it appears as though the neuron is more **"bursty"** during left and right turn trials as there is not consistency between trials for either turn type. In trial one, the left turns appear to be more regularly firing as they are evenly spaced out gaps in the data. However, this is not consistent with the trial 2 results for the left turn. For the right turns, the spike train appears to be "bursty" and inconsistent within both trial 1 and 2 as the breaks in firing is not consistent over the 30 second interval.

```python
# Calculate Inter-spike Interval (ISI)
ISI_left1 = np.diff(spikes_left_trial1)
ISI_left2 = np.diff(spikes_left_trial2)
ISI_right1 = np.diff(spikes_right_trial1)
ISI_right2 = np.diff(spikes_right_trial2)
```

```python
# Plot histograms of trials both left and right turns
f, a = plt.subplots(2, 2, figsize = (7,11))
bins = np.arange(0, 2, 0.01)
a[0][0].hist(ISI_left1, bins)
a[0][0].set_title('Trial 1: Left')
a[0][0].set_xlabel('ISI (s)')
a[0][0].set_ylabel('counts')

a[1][0].hist(ISI_left2, bins)
a[1][0].set_title('Trial 2: Left')
a[1][0].set_xlabel('ISI (s)')
a[1][0].set_ylabel('counts')

a[0][1].hist(ISI_right1, bins)
a[0][1].set_title('Trial 1: Right')
a[0][1].set_xlabel('ISI (s)')
a[0][1].set_ylabel('counts')

a[1][1].hist(ISI_right2, bins)
a[1][1].set_title('Trial 2: Right')
a[1][1].set_xlabel('ISI (s)')
a[1][1].set_ylabel('counts')
plt.show()
```

### Interpret Histograms

The neuron appears to fire in bursts however the ISI values are more contrained to lower ISI values between 0 and 0.25 s. As well as this, they possess long-ish tails going to bins as large as 1.5 seconds. Due to this, it confirms the bursting nature of the neuron. From my understanding, the Left Turn has higher burstiness as it is more spread out over the bin range than the Right which indicates greater irregularity. Overall, there appears to be variation between each trial and turn which is in-line with the spike train observations.

```python
# Calculating and Plotting the firing rate across time for each trial and turn type
binsize = 1
fr_bins = np.arange(0, 30, binsize)

# Initializing
fr_left1 = np.zeros(len(fr_bins)-1)
fr_left2 = np.zeros(len(fr_bins)-1)
fr_right1 = np.zeros(len(fr_bins)-1)
fr_right2 = np.zeros(len(fr_bins)-1)

for i in range(len(fr_bins)-1):
  spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])
  true_values = spikes_left_trial1[spk_index]
  fr_left1[i] = (len(true_values))/binsize

for i in range(len(fr_bins)-1):
  spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])
  true_values = spikes_left_trial2[spk_index]
  fr_left2[i] = (len(true_values))/binsize

for i in range(len(fr_bins)-1):
  spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])
  true_values = spikes_right_trial1[spk_index]
  fr_right1[i] = (len(true_values))/binsize

for i in range(len(fr_bins)-1):
  spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])
  true_values = spikes_right_trial2[spk_index]
  fr_right2[i] = (len(true_values))/binsize

# Plotting firing rate for trial 1 and trial 2
f, a = plt.subplots(2, 1, figsize = (8,10))

# Trial 1
a[0].plot(fr_bins[:-1], fr_left1, 'r')
a[0].plot(fr_bins[:-1], fr_right1, 'b')
a[0].set_xlabel('time (s)')
a[0].set_ylabel('Firing Rate (Hz)')
a[0].set_title('Trial One')
a[0].legend(['Left', 'Right'])

# Trial 2
a[1].plot(fr_bins[:-1], fr_left2, 'r')
a[1].plot(fr_bins[:-1], fr_right2, 'b')
a[1].set_xlabel('time (s)')
a[1].set_ylabel('Firing Rate (Hz)')
a[1].set_title('Trial Two')
a[1].legend(['Left', 'Right'])
```

### Interpret Firing Rate Over Time
*For trial 1*:
For the left turn, the firing right peaks in the middle of the 30 second period (10 - 15 seconds). Whereas, for the right turn, the firing rate peaks between 5 and 10 seconds. This means that they both have the highest firing rates during the middle of the waiting period. However, the firing rate of the left turn is almost double as much as the firing rate for the right turn.

*For trial 2*:
The graph for trial 2 appears to be significantly different than for trial 1. The maximum for the left turn in this case occurs between 22 and 27 seconds at the end of the waiting period. While the maximum for the right turn occurs in the beginning of the waiting period between 0 and 5 seconds. This means that during the second trial the greatest firing rate for the neuron occurs at an extreme of the waiting period. Another observation is that in this trial, the maximum right turn firing rate is about double the maximum left firing rate which is the opporside of what was previously observed.

*General*:
There does not appear to be any consistency between the two trials or directions.
