```python
# Import modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import relevant files as numpy arrays

url_accfa = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/acc_fa_hw6.csv'
url_acchits = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/acc_hits_hw6.csv'
url_lcfa = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/lc_fa_hw6.csv'
url_lchits = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/lc_hits_hw6.csv'

acc_fa = np.loadtxt(url_accfa, delimiter = ',')
acc_hits = np.loadtxt(url_acchits, delimiter = ',')
lc_fa = np.loadtxt(url_lcfa, delimiter = ',')
lc_hits = np.loadtxt(url_lchits, delimiter = ',')
```

```python
# Print shape of various arrays

print('False alarms: ', 'acc - ', acc_fa.shape, ', lc - ', lc_fa.shape)
print('Hits: ', 'acc - ', acc_hits.shape, ', lc - ', lc_hits.shape)
```

```python
# Creating a time variable

time = np.linspace(0, 4, acc_hits.shape[1])
```

```python
# Determine the sampling rate using time and number of samples

samples = acc_hits.shape[1]
time_end = time[-1]
sampleRate = samples/time_end

print(sampleRate, 'Hz')
```

```python
# Plot the first trial of hits and false alarms in subplots

f, a = plt.subplots(2, 1, figsize=(12, 3*3))

a[0].plot(time, acc_hits[0], 'r')
a[0].plot(time, lc_hits[0], 'b')
a[0].set_xlabel('time (s)')
a[0].set_ylabel('Voltage ($\mu$V)')
a[0].legend(['Acc', 'Lc'])

a[1].plot(time, acc_fa[0], 'r')
a[1].plot(time, lc_fa[0], 'b')
a[1].set_xlabel('time (s)')
a[1].set_ylabel('Voltage ($\mu$V)')
a[1].legend(['Acc', 'Lc'])
```

### Comparing LC and ACC for trial one
For the **ACC LFP**, the dominant oscillation appears to be between 1.3 and 2 Hz as that is the frequency between the highest amplitude peaks for hits and false alarms respectively. However, for the **LC LFP**, the dominant oscillation appears to be greaters as the highest amplitude peaks are closer together. For the hits and false alarms respectively the dominant oscillations appear to be 10 Hz and 5 Hz. In both cases the dominant oscillation is different.

Overall, there does not seem to be any visual correlation between the ACC and LC LFP results for either case. During the "*hit*", the peaks for the LC tend to occur at the same general timeframe as they do for the ACC. However, they are not at the same magnitude and do not occur at every peak. This trend is more consistent for the "*False Alarm*" case as the peaks for LC seem to occur after the peaks for the ACC with a slight lag. This is in line with what was discussed for phase coherence. However, it is most obvious for the false alarm case.



```python
# Compute the coherence for the acc and lc brain regions during a hit

dt = time[1] - time[0]
N = acc_hits.shape[1]
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real)                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

### Explaining the results

Coherence is over a range from 0 to 1; the closer the coherence is to 1 the more coherence between the two electrodes there are. This means that there would be less of a lag between the two sets of data. Therefore, in this set of data there is greater coherence at lower frequencies as the coherence is closer to one. In specific, the greatest coherence occurs at approximately **1 Hz** for the hit condition.

```python
# Calculating coherence for the false alarms and plotting it against hits

  # Hits
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real, 'r')                      # Plot coherence vs frequency,

  # False Alarms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real, 'b')                      # Plot coherence vs frequency,

plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
plt.legend(['hits', 'false alarms'])
```

### Coherence: hits vs false alarms
The previously created figure, shows that the frequency at which coherence is strongest is the same for both false alarms and hits. This occurs at approximately 1 Hz (decided visually from the graph). However, the overall coherence for the *false alarms* is lower than that for *hits*. This means that during hits the two brain areas, LC and ACC, are communicating with each other more than during the false alarms.
