````python
# Import matplotlib

from matplotlib import pyplot as plt

import pandas as pd

import numpy as np
````

````python
# Importing data files as Pandas dataframes

sst_url = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/v_rest_sst.csv'
pandas_data1 = pd.read_csv(sst_url)

pvalb_url = 'https://raw.githubusercontent.com/GraceHerchenroder/Neuro265/main/homework/v_rest_pvalb.csv'
pandas_data2 = pd.read_csv(pvalb_url)
````

````python
# Converting dataframes into a list

data1 = pandas_data1.values.tolist()
data2 = pandas_data2.values.tolist()
````

````python
# Eliminating the brackets around items in the list

data1 = np.squeeze(data1)
data2 = np.squeeze(data2)
````

````python
# Print length of data

print(len(data1))
````

````python
# Print the 3rd, 4th, 5th, and 6th elements of dataset 2

print(data2[2:6])
````

````python
# Print the second to last and last elements of dataset 1

print(data1[-2:])
````

````python
# Prints the type of dataset 2

type(data2)
````

````python
# First attempt at creating a histogram

plt.hist(data1)
````

````python
# Recreating histogram with axes labels

plt.hist(data1)
plt.xlabel('Resting Membrane Potential (V)')
plt.ylabel('Frequency')
plt.title('SST')
````

````python
# Creating histogram for dataset 2

plt.hist(data2)
plt.xlabel('Resting Membrane Potential (V)')
plt.ylabel('Frequency')
plt.title('Pvalb')
````

### **Data 1 versus Data 2**

The resting membrane potential of the Pvalb neurons tend to be slightly lower than that of the SST neurons. However, the two datasets occur over approximately the same range of resting membrane potentials. One of the more prominent differeces between the two histograms is that the Pvalb data has a higher frequency rand for up to 140 times as opposed to 120. This means that the Pvalb data is more concentrate toward the center (~ -70 V) of its range while the SST has a greater dispersion.
